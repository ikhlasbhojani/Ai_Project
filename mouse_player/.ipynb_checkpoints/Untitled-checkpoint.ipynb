```python
import pyautogui as pag
import mouse as m
import keyboard as kb
import time
import threading

def clicked():
    if m.is_pressed(button="left"):
        m.release(button="left")
        return True


def mouseRecord():
    while True:
        current_position = pag.position()
        if current_position[0] == 0 and current_position[1] == 0:
            time.sleep(0.1)
        elif current_position == pag.position():
            current_position = pag.position()
        else:
            path = pag.position()
            f = open("path.txt", "a")
            f.write(str(path[0]) + "," + str(path[1]) + "\n")
            f.close()
        if clicked() == True:
            f = open("path.txt", "a")
            f.write("Clicked\n")
            f.close()


def keybordRecord():
    def on_key_event(event):
        if event.event_type == "up":
            f = open("path.txt", "a")
            f.write(f"{event.name}\n")
            f.close()

    kb.hook(on_key_event)


def record():
    mouse_thread = threading.Thread(target=mouseRecord)
    keyboard_thread = threading.Thread(target=keybordRecord)

    keyboard_thread.start()
    mouse_thread.start()




def play():
    f = open("path.txt", "r")
    path = f.readlines()
    f.close()
    for i in range(len(path)):
        if path[i] == "Clicked\n":
            pag.click()
        elif any(c.isalpha() for c in path[i]):
            if path[i] == "enter\n":
                ent = path[i].split("\n")
                pag.press(ent)
            elif path[i] == "backspace\n":
                bs = path[i].split("\n")
                pag.press(bs)
            elif path[i] == "space\n":
                sp = path[i].split("\n")
                pag.press(sp)
            else:
                val = path[i].split("\n")
                pag.write(val)
        else:
            point = path[i]
            x = point.split(",")[0]
            y = point.split(",")[1]
            pag.moveTo(int(x), int(y))
# record()
play()
```
